```python
diffMIR = OrderedDict()
allMIR = OrderedDict()
with open('diffmir.txt') as f:
    for line in f:
        line = line.rstrip()
        lst = line.split('\t')
        if lst[0] == 'miR-name':
            continue
        else:
            diffMIR[lst[0]] = lst[-1]
            
            
fh1 = open('allmirseq.clean.txt', 'wt')
fh2 = open('all_mir.fa', 'wt')
fh3 = open('diff_mir.fa', 'wt')

test = OrderedDict()        
with open('allmirseq.txt') as f:
    for line in f:
        line = line.rstrip()
        lst= line.split('\t')
        if lst[0] == 'miRNA_name':
            fh1.write(line+'\n')
            continue
        elif lst[0] in allMIR.keys():
            continue
        
        allMIR[lst[0]] = lst[1]
        fh1.write(line+'\n')
        
        fh2.write('>'+lst[0]+'\n'+lst[1]+'\n')
        
        if lst[0] in diffMIR.keys():
            test[lst[0]] = 1
            fh3.write('>'+lst[0]+'\n'+lst[1]+'\n')
            
            
fh1.close()
fh2.close()
fh3.close()
```

```python
degradome= []
pstarget = []
overlaps = []


fh1 = open('degradome_target.clean.txt', 'wt')
with open('dedradome_target.txt') as f:
    for line in f:
        if line.startswith('miRNA'):
            continue
        line = line.rstrip()
        lst = line.split('\t')
        
        lst[1] = re.sub(r'(\d+)g', r'\1G', lst[1])
        
        miRNAs = re.split(',| ', lst[0])
        
        for mir in miRNAs:
            if mir == '':
                continue

            if mir+'\t'+lst[1] not in degradome:
                degradome.append(mir+'\t'+lst[1])
                fh1.write(mir+'\t'+lst[1]+'\n')
                
fh1.close()

        
                
with open('psRNATarget_diffmir_JGI.txt') as f:
    for line in f:
        if line.startswith('#') or line.startswith('miRNA'):
            continue
        line = line.rstrip()
        lst = line.split('\t')
       
        if lst[0]+'\t'+lst[1] not in pstarget:
            pstarget.append(lst[0]+'\t'+lst[1])

                      
fh = open('degradome_pstarget_overlaps.txt', 'wt')
for target in degradome:
    if target in pstarget:
        overlaps.append(target)
        fh.write(target+'\n')
        
        
fh.close()
        
        
print ('number of interactions in %s: %d' %('degradome', len(degradome)))
print ('number of interactions in %s: %d' %('pstarget', len(pstarget)))

print ('number of interactions in %s: %d' %('overlaps', len(overlaps)))
```

```python
##########################
###### GO

goBasic = OrderedDict()
with open('go-basic-02142017.obo.txt') as f:
    for line in f:
        line = line.rstrip()
        lst = line.split('\t')
        goBasic[lst[1]] = lst[2]
        
goBasic['NA'] = 'NA'


goBP = OrderedDict()
goMF = OrderedDict()
goCC = OrderedDict()

pfam = OrderedDict()
panther = OrderedDict()
kog = OrderedDict()


def smry_anno_go(lst,goBP, j=11):
    bpID = lst[j].split(' ')
    bpSmry = [k+'\t'+goBasic[k] if k in goBasic.keys() else k+'\tNA' for k in bpID]
        
    for bp in bpSmry:
        if bp not in goBP.keys():
            goBP[bp] = []
        goBP[bp].append(lst[0])


def smry_anno(lst,pfam, j=17):
    bpID = lst[j].split(' ')
    bpTerm = lst[j+1].split('; ')
    bpSmry = [bpID[i]+'\t'+bpTerm[i] for i in range(len(bpID))]
        
    for bp in bpSmry:
        if bp not in pfam.keys():
            pfam[bp] = []
        pfam[bp].append(lst[0])


with open('summary.degradome_target_diffmir.soybase.anno.txt') as f:
    for line in f:
        line = line.rstrip()
        line = line.replace('"', '')
        lst = line.split('\t')
       
        if line.startswith('Gmax'):
            continue
            
        smry_anno_go(lst,goBP, j=11)
        smry_anno_go(lst,goMF, j=13)
        smry_anno_go(lst,goCC, j=15)
        smry_anno(lst,pfam, j=17)
        smry_anno(lst,panther, j=19)
        smry_anno(lst,kog, j=21)
        
        
        goBPsort = [(ke, goBP[ke]) for ke in sorted(goBP, key=lambda k: len(goBP[k]), reverse=True)]
        goMFsort = [(ke, goMF[ke]) for ke in sorted(goMF, key=lambda k: len(goMF[k]), reverse=True)]
        goCCsort = [(ke, goCC[ke]) for ke in sorted(goCC, key=lambda k: len(goCC[k]), reverse=True)]
        pfamsort = [(ke, pfam[ke]) for ke in sorted(pfam, key=lambda k: len(pfam[k]), reverse=True)]
        panthersort = [(ke, panther[ke]) for ke in sorted(panther, key=lambda k: len(panther[k]), reverse=True)]
        kogsort = [(ke, kog[ke]) for ke in sorted(kog, key=lambda k: len(kog[k]), reverse=True)]
        
        
fh = open('deg_diffmir_target_soybase_anno.goBP.txt', 'wt')
for ke, val in goBPsort:
    fh.write(ke+'\t'+str(len(val))+'\t'+';'.join(val)+'\n')
    
fh.close()


fh = open('deg_diffmir_target_soybase_anno.goMF.txt', 'wt')
for ke, val in goMFsort:
    fh.write(ke+'\t'+str(len(val))+'\t'+';'.join(val)+'\n')
    
fh.close()


fh = open('deg_diffmir_target_soybase_anno.goCC.txt', 'wt')
for ke, val in goCCsort:
    fh.write(ke+'\t'+str(len(val))+'\t'+';'.join(val)+'\n')
    
fh.close()


fh = open('deg_diffmir_target_soybase_anno.PFAM.txt', 'wt')
for ke, val in pfamsort:
    fh.write(ke+'\t'+str(len(val))+'\t'+';'.join(val)+'\n')
    
fh.close()


fh = open('deg_diffmir_target_soybase_anno.Panther.txt', 'wt')
for ke, val in panthersort:
    fh.write(ke+'\t'+str(len(val))+'\t'+';'.join(val)+'\n')
    
fh.close()


fh = open('deg_diffmir_target_soybase_anno.KOG.txt', 'wt')
for ke, val in kogsort:
    fh.write(ke+'\t'+str(len(val))+'\t'+';'.join(val)+'\n')
    
fh.close()
```
